```python
df_clean['tot_sale']=df_clean.Quantity*df_clean.UnitPrice
```

```python
#8) Find how many different customers are available for each country save the count
#in dictionary also return countries with highest and lowest customer count

my_country=set(df_clean1['Country'])
countr_cust_dict={}
for i in my_country:
  no_cust=len(df_clean1[df_clean1['Country']==i]['CustomerID'].unique())
  countr_cust_dict[i]=no_cust
```

```python
coust_lst=[(value, key) for key, value in countr_cust_dict.items()]
coust_lst
```

```python
print(max(countr_cust_dict.values()),max(countr_cust_dict, key=countr_cust_dict.get))
```

```python
#9) Find which stock code was ordered the most by each country
df_clean1[df_clean1['Country']=='United Kingdom'][['StockCode','Quantity']].groupby(['StockCode']).sum().max()[]
```

```python
max_stock_dict={}
for i in my_country:
  stock_name=df_clean1[df_clean1['Country']==i][['StockCode','Quantity']].groupby(['StockCode']).sum().idxmax()
  quantity_of_stock=df_clean1[df_clean1['Country']==i][['StockCode','Quantity']].groupby(['StockCode']).sum().max()
  max_stock_dict[i]=[stock_name['Quantity'],quantity_of_stock['Quantity']]
```
